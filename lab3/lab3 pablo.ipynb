### Hopefield network implementation

```python
import numpy as np
import matplotlib.pyplot as plt
from hopfield import HopfieldNetwork
```

```python
def visualize_binary_pattern(pattern):
    pattern_size = int(np.sqrt(len(pattern)) if np.sqrt(len(pattern)) % 1 == 0 else len(pattern))
    pattern_image = np.reshape(pattern, (pattern_size, pattern_size))

    plt.imshow(pattern_image, cmap='binary')
    plt.title('Generated Binary Pattern')
    plt.axis('off')  # Hide axis labels
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

activity = 0.01
# Create a Hopfield network with a specified number of neurons
nb_neurons = 100

# Function to generate sparse binary patterns with 10% activity
def generate_sparse_patterns(num_patterns, pattern_size):
    patterns = np.random.choice([0, 1], size=(num_patterns, pattern_size), p=[1-activity,activity])
    return patterns

# Function to test the network for different values of Î¸
def test_hopfield_network(patterns, theta_values):
    stored_patterns = []

    for bias in theta_values:
        network = HopfieldNetwork(max_iterations=10, sparsity=activity, bias=bias)

        network.fit(patterns, patterns)
        correct_patterns = 0
        print(f"New theta is: {bias}")

        for pattern in patterns:

            reconstruction = network.predict(pattern)
            if (np.array_equal(pattern, reconstruction)):
                correct_patterns += 1

        stored_patterns.append(correct_patterns)

    return stored_patterns


def experiment1():
    num_patterns = 300
    pattern_size = nb_neurons
    theta_values = [i * 0.01 for i in range(41)]  # Adjust theta values as needed

    patterns = generate_sparse_patterns(num_patterns, pattern_size)
    # for pattern in patterns:
    #     visualize_binary_pattern(pattern)
    stored_patterns = test_hopfield_network(patterns, theta_values)
    print(stored_patterns)
    # Plot the results
    plt.ylim(0, 300)
    plt.plot(theta_values, stored_patterns, marker='o')
    plt.xlabel('Theta (Bias)')
    plt.ylabel('Number of Stored Patterns')
    plt.title('Hopfield Network Storage Capacity')
    plt.grid()
    plt.show()
```

```python
experiment1()
```
